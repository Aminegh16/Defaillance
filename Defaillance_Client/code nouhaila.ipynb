```python
import numpy as np

# Ajouter la colonne "Ratio Epuisé" en gérant la division par zéro
df_3_5['Ratio Epuisé'] = np.where(df_3_5['CA 23'] != 0, np.abs(df_3_5['Mtant 23'] / df_3_5['CA 23']), np.nan)

df_3_5
```

```python
# Fonction de classification pour CA 23
def classify_ca(ca):
    if ca >= 100e6:
        return 'A'
    elif 10e6 <= ca < 100e6:
        return 'B'
    elif 1e6 <= ca < 10e6:
        return 'C'
    elif 100e3 <= ca < 1e6:
        return 'D'
    else:
        return 'E'

# Ajouter la colonne de classification pour CA 23
df_3_5['Classe CA 23'] = df_3_5['CA 23'].apply(classify_ca)

# Trier les données par CA 23 en ordre décroissant
df_3_5_sorted = df_3_5.sort_values(by='CA 23', ascending=False)
```

```python
# Fonction de classification pour Mtant 23
def classify_mtant(mtant):
    mtant_abs = abs(mtant)
    if mtant_abs >= 100e6:
        return 'A'
    elif 10e6 <= mtant_abs < 100e6:
        return 'B'
    elif 1e6 <= mtant_abs < 10e6:
        return 'C'
    elif 100e3 <= mtant_abs < 1e6:
        return 'D'
    else:
        return 'E'

# Ajouter la colonne de classification pour Mtant 23
df_3_5['Classe Mtant 23'] = df_3_5['Mtant 23'].apply(classify_mtant)

print(df_3_5[['Mtant 23', 'Classe Mtant 23']])

# Ajouter la colonne de classification pour CA 23
df_3_5['Classe CA 23'] = df_3_5['CA 23'].apply(classify_ca)

# Trier les données par CA 23 en ordre décroissant
df_3_5_sorted = df_3_5.sort_values(by='CA 23', ascending=False)

# Afficher les colonnes souhaitées
df_3_5_sorted
```
